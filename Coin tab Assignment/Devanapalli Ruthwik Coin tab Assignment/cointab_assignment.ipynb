```python
# importing all important packages
import pandas as pd
```

```python
# Define the file path
file_path = 'Final Assignment Data/Company X - Order Report.xlsx'

# Load the Excel file into a pandas DataFrame
Company_X_Order_Report = pd.read_excel(file_path)

# Now, 'data' contains your Excel file's data
Company_X_Order_Report.head()
```

```python
# Load Company X - Pincode Zones data into a pandas DataFrame from an Excel file

# Define the file path
file_path = 'Final Assignment Data/Company X - Pincode Zones.xlsx'

# Load the Excel file into a pandas DataFrame
Company_X_Pincode_Zones = pd.read_excel(file_path)

# Now, 'data' contains your Excel file's data
Company_X_Pincode_Zones.head()
```

```python
# Load Company X - SKU Master data into a pandas DataFrame from an Excel file
file_path = 'Final Assignment Data/Company X - SKU Master.xlsx'
Company_X_SKU_Master = pd.read_excel(file_path)
Company_X_SKU_Master.head()
```

```python
# Load Courier Company - Invoice data into a pandas DataFrame from an Excel file
file_path = 'Final Assignment Data/Courier Company - Invoice.xlsx'
Courier_Company_Invoice = pd.read_excel(file_path)
Courier_Company_Invoice.head()
```

```python
# Load Courier Company - Rates data into a pandas DataFrame from an Excel file
file_path = 'Final Assignment Data/Courier Company - Rates.xlsx'
Courier_Company_Rates= pd.read_excel(file_path)
Courier_Company_Rates.head()
```

```python
Courier_Company_Rates['Zone'] = Courier_Company_Rates['Zone'].str.lower()
Courier_Company_Rates.head()
```

```python
merged_data = pd.merge(Courier_Company_Invoice,Company_X_Pincode_Zones, on=['Warehouse Pincode', 'Customer Pincode'], how='inner')
merged_data.head()
```

```python
merged_data = pd.merge(merged_data,Company_X_Order_Report, left_on='Order ID', right_on='ExternOrderNo', how='right')
merged_data.head()
```

```python
merged_data = pd.merge(merged_data,Company_X_SKU_Master, on='SKU')
merged_data.head()
```

```python
merged_data = pd.merge(merged_data,Courier_Company_Rates, left_on='Zone_x', right_on='Zone', how='inner')
merged_data.head()
```

```python
# Convert 'Weight (g)' to kilograms
merged_data['Weight (kg)'] = merged_data['Weight (g)'] / 1000
merged_data.head()
```

```python
merged_data.drop(["ExternOrderNo", "Weight (g)","Zone"], axis=1, inplace=True)
merged_data.rename(columns={'Zone_x': 'Delivery Zone charged by Courier Company', 'Zone_y': 'Delivery Zone as per X','AWB Code': 'AWB Number','Billing Amount (Rs.)': 'Charges Billed by Courier Company (Rs.)','Weight (kg)': 'Total weight as per X (KG)','Weight Slabs' :'Weight slab charged by Courier Company (KG)','Charged Weight' :'Total weight as per Courier Company (KG)'}, inplace=True)
merged_data.head()
```

```python
# Function to calculate charges
def calculate_charges(row):
    zone = row['Delivery Zone as per X']
    weight = row['Total weight as per X (KG)']

    # Check if zone exists in Courier_Company_Rates
    if zone not in Courier_Company_Rates['Zone'].unique():
        return 0  # or any default value

    weight_slab = Courier_Company_Rates[(Courier_Company_Rates['Zone'] == zone) & (weight <= Courier_Company_Rates['Weight Slabs'])]

    # Check if weight_slab is empty
    if weight_slab.empty:
        return 0  # or any default value

    # Ensure weight_slab has at least one row before accessing the first row
    if not weight_slab.index.empty:
        weight_slab = weight_slab.iloc[0]  # Select the first row

        fixed_charge = weight_slab['Forward Fixed Charge']
        additional_charge = weight_slab['Forward Additional Weight Slab Charge']

        total_weight = weight
        total_additional_weight = max(0, total_weight - weight_slab['Weight Slabs'])
        total_charge = fixed_charge + (total_additional_weight * additional_charge)

        return total_charge
    else:
        return 0  # or any default value if weight_slab is empty
```

```python
# Apply the calculate_charges function to the merged_data dataframe
merged_data['Expected Charge as per X (Rs.)'] = merged_data.apply(calculate_charges, axis=1)
merged_data.head()
```

```python
# Create the output data
output_data = merged_data[['Order ID', 'AWB Number', 'Total weight as per X (KG)',
                            'Total weight as per Courier Company (KG)',
                            'Weight slab charged by Courier Company (KG)', 'Delivery Zone as per X',
                            'Delivery Zone charged by Courier Company', 'Expected Charge as per X (Rs.)',
                            'Charges Billed by Courier Company (Rs.)']]

# Rename the columns
output_data.columns = ['Order ID', 'AWB Number', 'Total weight as per X (KG)',
                        'Total weight as per Courier Company (KG)', 'Weight slab charged by Courier Company (KG)',
                        'Delivery Zone as per X', 'Delivery Zone charged by Courier Company',
                        'Expected Charge as per X (Rs.)', 'Charges Billed by Courier Company (Rs.)']

# Display the first few rows of the output_data DataFrame
output_data.head()
```

```python
# Calculate the difference
output_data['Difference Between Expected Charges and Billed Charges (Rs.)'] = output_data['Expected Charge as per X (Rs.)'] - output_data['Charges Billed by Courier Company (Rs.)']
```

```python
# Save the output data to a CSV file
output_data.to_csv("output_data.csv", index=False)
```

```python
# Create the summary table
summary_table = pd.DataFrame({
    'Total orders where X has been correctly charged': [output_data['Difference Between Expected Charges and Billed Charges (Rs.)'].eq(0).sum()],
    'Total Orders where X has been overcharged': [output_data['Difference Between Expected Charges and Billed Charges (Rs.)'].gt(0).sum()],
    'Total Orders where X has been undercharged': [output_data['Difference Between Expected Charges and Billed Charges (Rs.)'].lt(0).sum()],
    'Total invoice amount': [output_data['Charges Billed by Courier Company (Rs.)'].sum()],
    'Total overcharging amount': [output_data['Difference Between Expected Charges and Billed Charges (Rs.)'].where(lambda x: x > 0, 0).sum()],
    'Total undercharging amount': [output_data['Difference Between Expected Charges and Billed Charges (Rs.)'].where(lambda x: x < 0, 0).sum()]
})

summary_table.head()
```

```python
# Save the summary table to a CSV file
summary_table.to_csv("summary_table.csv", index=False)
```
