```python


# Merge datasets
merged_data = pd.merge(Company_X_Order_Report,Company_X_Pincode_Zones, on='Warehouse Pincode')
merged_data = pd.merge(merged_data, sku_master, on='SKU')
merged_data = pd.merge(merged_data, courier_invoice, on='Order ID')

# Function to calculate applicable weight based on zone's weight slab
def calculate_applicable_weight(row):
    zone = row['Delivery Zone']
    weight = row['Total Weight (KG)']
    slab = courier_rate_card[(courier_rate_card['Zone'] == zone)]['Slab (KG)'].iloc[0]
    return min(weight, slab)

# Apply function to calculate applicable weight
merged_data['Applicable Weight'] = merged_data.apply(calculate_applicable_weight, axis=1)

# Create a function to calculate expected charge
def calculate_expected_charge(row):
    weight = row['Total Weight (KG)']
    zone = row['Delivery Zone']
    slab = courier_rate_card[(courier_rate_card['Zone'] == zone)]['Slab (KG)'].iloc[0]
    fixed_charge = courier_rate_card[(courier_rate_card['Zone'] == zone)]['Fixed'].iloc[0]
    additional_charge = courier_rate_card[(courier_rate_card['Zone'] == zone)]['Additional'].iloc[0]
    total_additional = max(0, weight - slab)
    return fixed_charge + total_additional * additional_charge

# Apply function to calculate expected charge
merged_data['Expected Charge (Rs.)'] = merged_data.apply(calculate_expected_charge, axis=1)

# Generate the resultant dataframe
result_df = merged_data[['Order ID', 'AWB Number', 'Total Weight (KG)', 'Weight Slab (KG)', 
                         'Total Weight (KG)_x', 'Weight Slab Charged (KG)', 'Delivery Zone', 
                         'Delivery Zone Charged', 'Expected Charge (Rs.)', 'Charges (Rs.)']]
result_df['Difference (Rs.)'] = result_df['Expected Charge (Rs.)'] - result_df['Charges (Rs.)']

# Create summary table
summary_table = pd.DataFrame({
    'Count': [len(result_df),
              len(result_df[result_df['Difference (Rs.)'] == 0]),
              len(result_df[result_df['Difference (Rs.)'] > 0]),
              len(result_df[result_df['Difference (Rs.)'] < 0])],
    'Amount (Rs.)': [result_df['Charges (Rs.)'].sum(),
                     result_df[result_df['Difference (Rs.)'] == 0]['Charges (Rs.)'].sum(),
                     result_df[result_df['Difference (Rs.)'] > 0]['Difference (Rs.)'].sum(),
                     result_df[result_df['Difference (Rs.)'] < 0]['Difference (Rs.)'].sum()]
}, index=['Total Orders', 'Correctly Charged', 'Overcharged', 'Undercharged'])

# Save the results to Excel
with pd.ExcelWriter('result.xlsx') as writer:
    result_df.to_excel(writer, sheet_name='Order Level Calculation', index=False)
    summary_table.to_excel(writer, sheet_name='Summary Table')

print("Script executed successfully. Results saved to result.xlsx")
```
